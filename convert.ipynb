```python
import nibabel as nib
import numpy as np

# Load your aseg file
seg = nib.load(r"C:\Users\Aadya\OneDrive\Desktop\Brain\aseg.nii.gz")
seg_data = seg.get_fdata().astype(int)

# Create an empty mask with same shape
mask = np.zeros_like(seg_data, dtype=np.uint8)

# Example grouping (adjust if needed)
gray_matter = [3, 42]                 # left & right cortex
white_matter = [2, 41]                # left & right white matter
csf = [4, 43, 14, 15]                 # ventricles and CSF spaces

# Assign new labels
mask[np.isin(seg_data, gray_matter)] = 1   # GM → 1
mask[np.isin(seg_data, white_matter)] = 2  # WM → 2
mask[np.isin(seg_data, csf)] = 3           # CSF → 3
# Background stays 0

# Save new mask
new_seg = nib.Nifti1Image(mask, seg.affine)
nib.save(new_seg, "aseg_simplified.nii.gz")

print("New mask saved as aseg_simplified.nii.gz")
```

```python
import os
import nibabel as nib
import numpy as np
```

```python
data_folder = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\images"  # replace with your dataset path
subjects = os.listdir(data_folder)

for sub in subjects:
    sub_path = os.path.join(data_folder, sub)

    # Brain MRI
    brain_file = os.path.join(sub_path, "brain.nii.gz")
    if os.path.exists(brain_file):
        brain = nib.load(brain_file)
        brain_data = brain.get_fdata()
        # normalize between 0–1
        brain_norm = (brain_data - np.min(brain_data)) / (np.max(brain_data) - np.min(brain_data))
        nib.save(nib.Nifti1Image(brain_norm, brain.affine), os.path.join(sub_path, "brain_normalized.nii.gz"))

    # Label (aseg simplified) – do NOT normalize
    label_file = os.path.join(sub_path, "aseg_simplified.nii.gz")
    if os.path.exists(label_file):
        label = nib.load(label_file)
        label_data = label.get_fdata().astype(np.int16)  # keep as integers
        nib.save(nib.Nifti1Image(label_data, label.affine), os.path.join(sub_path, "label_ready.nii.gz"))

print("✅ Preprocessing done for all subjects!")
```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import nibabel as nib
import numpy as np
```

```python
class BrainSegDataset(Dataset):
    def __init__(self, image_dir, label_dir, transform=None):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.image_files = sorted(os.listdir(image_dir))
        self.label_files = sorted(os.listdir(label_dir))
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        # Load brain image
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        img = nib.load(img_path).get_fdata()

        # Load mask
        label_path = os.path.join(self.label_dir, self.label_files[idx])
        label = nib.load(label_path).get_fdata()

        # Convert to float32
        img = img.astype(np.float32)
        label = label.astype(np.int64)  # keep as integers for segmentation

        # Add channel dimension (PyTorch expects [C, D, H, W])
        img = np.expand_dims(img, axis=0)

        # Convert to tensors
        img = torch.tensor(img, dtype=torch.float32)
        label = torch.tensor(label, dtype=torch.long)

        return img, label


# Paths
train_img_dir = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\train\images"
train_lbl_dir = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\train\labels"

# Dataset and DataLoader
train_dataset = BrainSegDataset(train_img_dir, train_lbl_dir)
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)

# Test loading one batch
for imgs, masks in train_loader:
    print("Image shape:", imgs.shape)
    print("Mask shape:", masks.shape)
    break
```

```python
import torch
import torch.nn as nn

# Building block: Convolution + BatchNorm + ReLU
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.double_conv = nn.Sequential(
            nn.Conv3d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm3d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv3d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm3d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

# U-Net
class UNet3D(nn.Module):
    def __init__(self, in_channels=1, out_channels=2):  # out_channels=2 for binary segmentation
        super(UNet3D, self).__init__()

        # Encoder (downsampling)
        self.enc1 = DoubleConv(in_channels, 32)
        self.pool1 = nn.MaxPool3d(2)
        self.enc2 = DoubleConv(32, 64)
        self.pool2 = nn.MaxPool3d(2)
        self.enc3 = DoubleConv(64, 128)
        self.pool3 = nn.MaxPool3d(2)

        # Bottleneck
        self.bottleneck = DoubleConv(128, 256)

        # Decoder (upsampling)
        self.up3 = nn.ConvTranspose3d(256, 128, kernel_size=2, stride=2)
        self.dec3 = DoubleConv(256, 128)
        self.up2 = nn.ConvTranspose3d(128, 64, kernel_size=2, stride=2)
        self.dec2 = DoubleConv(128, 64)
        self.up1 = nn.ConvTranspose3d(64, 32, kernel_size=2, stride=2)
        self.dec1 = DoubleConv(64, 32)

        # Final output
        self.final_conv = nn.Conv3d(32, out_channels, kernel_size=1)

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool1(e1))
        e3 = self.enc3(self.pool2(e2))

        # Bottleneck
        b = self.bottleneck(self.pool3(e3))

        # Decoder with skip connections
        d3 = self.up3(b)
        d3 = torch.cat((d3, e3), dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        d2 = torch.cat((d2, e2), dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        d1 = torch.cat((d1, e1), dim=1)
        d1 = self.dec1(d1)

        return self.final_conv(d1)


# Test the model
model = UNet3D(in_channels=1, out_channels=2)  # out_channels=2 → brain vs background
x = torch.randn(1, 1, 128, 128, 128)  # dummy input
y = model(x)

print("Input shape:", x.shape)
print("Output shape:", y.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torch.nn.functional as F

# ---------- Dice Loss for binary segmentation ----------
class DiceLoss(nn.Module):
    def __init__(self, smooth=1e-5):
        super(DiceLoss, self).__init__()
        self.smooth = smooth

    def forward(self, preds, targets):
        # preds = [B, C, D, H, W], targets = [B, D, H, W]
        preds = torch.sigmoid(preds)  # convert logits → probabilities (for binary)
        targets = targets.float()     # binary mask as float

        # Flatten
        preds_flat = preds.view(-1)
        targets_flat = targets.view(-1)

        intersection = (preds_flat * targets_flat).sum()
        dice = (2. * intersection + self.smooth) / (preds_flat.sum() + targets_flat.sum() + self.smooth)
        return 1 - dice  # we want to minimize

# ---------- Dataset Loader ----------
train_dataset = BrainSegDataset(
    image_dir=r"datasets/train/images",
    label_dir=r"datasets/train/labels"
)

train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)

# ---------- Model ----------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UNet3D(in_channels=1, out_channels=1).to(device)  # out_channels=1 for binary

optimizer = optim.Adam(model.parameters(), lr=1e-4)
num_epochs = 5  # adjust later

# ---------- Training Loop ----------
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0

    for images, masks in train_loader:
        images, masks = images.to(device), masks.to(device)

        # Convert masks to binary (brain vs background)
        masks = (masks > 0).long()  # everything >0 is brain

        # Forward pass
        outputs = model(images)
        loss = DiceLoss()(outputs, masks)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(train_loader):.4f}")

# ---------- Save Model ----------
torch.save(model.state_dict(), "unet3d_brain_binary.pth")
print("✅ Training complete. Model saved as unet3d_brain_binary.pth")
```

```python
# Step1_registration.py  -- run this cell only for Step 1
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib

# ==== PATHS (change these) ====
template_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\healthy_brain\mni_icbm152_t1_tal_nlin_asym_09a.nii.gz"   # pick a healthy brain template
subject_path  = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\test\images\OAS30002_MR_d2340_normalize_resized.nii"
subject_mask_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\test\labels\OAS30002_MR_d2340_mask_resized.nii"

out_subject_reg = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\test\OAS30010_registered.nii.gz"
out_mask_reg    = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\test\OAS30010_mask_registered.nii.gz"

# ==== Load images ====
template = sitk.ReadImage(template_path, sitk.sitkFloat32)
subject  = sitk.ReadImage(subject_path, sitk.sitkFloat32)
mask     = sitk.ReadImage(subject_mask_path, sitk.sitkUInt8)  # mask is integer

# ==== Rigid registration (centered) ====
initial_transform = sitk.CenteredTransformInitializer(template, subject,
                                                     sitk.Euler3DTransform(),
                                                     sitk.CenteredTransformInitializerFilter.GEOMETRY)

registration_method = sitk.ImageRegistrationMethod()
registration_method.SetMetricAsMattesMutualInformation(numberOfHistogramBins=50)
registration_method.SetMetricSamplingStrategy(registration_method.RANDOM)
registration_method.SetMetricSamplingPercentage(0.01)
registration_method.SetInterpolator(sitk.sitkLinear)
registration_method.SetOptimizerAsGradientDescent(learningRate=1.0, numberOfIterations=100,
                                                  convergenceMinimumValue=1e-6, convergenceWindowSize=10)
registration_method.SetOptimizerScalesFromPhysicalShift()
registration_method.SetInitialTransform(initial_transform, inPlace=False)
registration_method.SetShrinkFactorsPerLevel(shrinkFactors = [4,2,1])
registration_method.SetSmoothingSigmasPerLevel(smoothingSigmas=[2,1,0])
registration_method.SmoothingSigmasAreSpecifiedInPhysicalUnitsOn()

final_transform = registration_method.Execute(template, subject)

print("Optimizer stop condition: {0}".format(registration_method.GetOptimizerStopConditionDescription()))
print("Final metric value: {0}".format(registration_method.GetMetricValue()))

# ==== Resample subject image into template space ====
resampled_subject = sitk.Resample(subject, template, final_transform, sitk.sitkLinear, 0.0, subject.GetPixelID())
resampled_mask    = sitk.Resample(mask,    template, final_transform, sitk.sitkNearestNeighbor, 0, mask.GetPixelID())

# ==== Save outputs ====
sitk.WriteImage(resampled_subject, out_subject_reg)
sitk.WriteImage(resampled_mask, out_mask_reg)

print("Saved registered subject and mask to:")
print(out_subject_reg)
print(out_mask_reg)

# ==== Quick visual check (middle slice) ====
sub_np = sitk.GetArrayFromImage(resampled_subject)  # shape: [Z,Y,X]
mask_np = sitk.GetArrayFromImage(resampled_mask)

z_mid = sub_np.shape[0] // 2
plt.figure(figsize=(8,8))
plt.imshow(sub_np[z_mid,:,:], cmap='gray')
plt.imshow(mask_np[z_mid,:,:], cmap='Reds', alpha=0.35)
plt.title("Registered subject (gray) + mask overlay (red) — slice {}".format(z_mid))
plt.axis('off')
plt.show()
```

```python
import numpy as np
import pandas as pd
from nilearn import datasets
from nilearn.image import load_img, resample_to_img

# === Load your mask ===
mask_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\test\OAS30010_mask_registered.nii.gz"
mask_img = load_img(mask_path)

# === Load Harvard-Oxford atlas ===
atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-2mm')
atlas_img = load_img(atlas.maps)
atlas_labels = atlas.labels

# === Resample atlas to match mask dimensions ===
atlas_resampled = resample_to_img(atlas_img, mask_img, interpolation='nearest')

# === Convert images to arrays ===
atlas_data = atlas_resampled.get_fdata().astype(int)
mask_data = mask_img.get_fdata() > 0   # Create diff_mask from your mask
diff_mask = mask_data.astype(bool)

# === Quantify affected regions ===
region_voxels = {}

for region_id, region_name in enumerate(atlas_labels):
    if region_id == 0:  # Skip background
        continue
    region_mask = (atlas_data == region_id)
    affected_count = np.sum(diff_mask[region_mask])
    total_count = np.sum(region_mask)
    if total_count > 0:
        affected_percent = (affected_count / total_count) * 100
        if affected_count > 0:  # Include only affected regions
            region_voxels[region_name] = [affected_count, affected_percent]

# === Convert to DataFrame ===
df = pd.DataFrame(region_voxels, index=["Affected Voxels", "Percent (%)"]).T
df_sorted = df.sort_values(by="Percent (%)", ascending=False)

# === Display top 15 affected regions ===
print(df_sorted.head(15))

# === Save to CSV ===
df_sorted.to_csv("affected_regions_summary.csv", index=True)
print("\n✅ Saved as 'affected_regions_summary.csv'")
```

```python
import os
import nibabel as nib
import numpy as np
import torch
from torch.utils.data import Dataset

class MRIDataset(Dataset):
    def __init__(self, folder, csv_file, transform=None):
        """
        folder: path to images folder
        csv_file: path to matched_preprocessed_sessions.csv
        """
        self.folder = folder
        self.df = pd.read_csv(csv_file)
        self.transform = transform
        
        # Create mapping: session_id -> label
        self.label_map = dict(zip(self.df['mri_session_id'], self.df['diagnosis']))
        
        # List of image files
        self.files = [f for f in os.listdir(folder) if f.endswith(".nii") or f.endswith(".nii.gz")]

    def __len__(self):
        return len(self.files)
    
    def __getitem__(self, idx):
        file_name = self.files[idx]
        path = os.path.join(self.folder, file_name)
        
        # Load .nii image
        img = nib.load(path).get_fdata()
        
        # Take middle slice
        slice_img = img[:, :, img.shape[2]//2]
        
        # Resize to 224x224
        slice_img = np.array(np.resize(slice_img, (224, 224)), dtype=np.float32)
        
        # Normalize
        slice_img = (slice_img - slice_img.min()) / (slice_img.max() - slice_img.min())
        
        # Add channel dimension
        slice_img = np.expand_dims(slice_img, axis=0)
        
        # Get label
        session_id = file_name.split("_Raw")[0]  # adjust if needed
        label = self.label_map.get(session_id)
        
        # Convert label to integer (optional, e.g., 0=Control, 1=MCI, 2=AD)
        classes = sorted(self.df['diagnosis'].unique())
        label_map_int = {cls:i for i, cls in enumerate(classes)}
        label_int = label_map_int[label]
        
        # Convert to torch tensors
        img_tensor = torch.tensor(slice_img, dtype=torch.float32)
        label_tensor = torch.tensor(label_int, dtype=torch.long)
        
        return img_tensor, label_tensor
```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import nibabel as nib
import pandas as pd
import numpy as np

class MRIDataset(Dataset):
    def __init__(self, images_dir, csv_file):
        """
        images_dir: folder containing MRI .nii files
        csv_file: CSV with columns:
                  mri_session_id, diagnosis
        """
        self.images_dir = images_dir
        self.df = pd.read_csv(csv_file)
        
        # Map diagnosis to integers
        self.label_map_int = {"Control": 0, "MCI": 1, "Alzheimer": 2}
        
        # Build a mapping: session_id -> label
        self.label_map = {row['mri_session_id']: row['diagnosis'] for idx, row in self.df.iterrows()}
        
        # List of image files
        self.image_files = [f for f in os.listdir(images_dir) if f.endswith(".nii") or f.endswith(".nii.gz")]
    
    def __len__(self):
        return len(self.image_files)
    
    def __getitem__(self, idx):
        # Get filename
        img_file = self.image_files[idx]
        session_id = img_file.replace("_normalize_resized.nii", "").replace(".nii.gz", "")
        
        # Load image
        img_path = os.path.join(self.images_dir, img_file)
        img_nib = nib.load(img_path)
        img = img_nib.get_fdata()
        
        # Convert to float32 and add channel dimension
        img_tensor = torch.tensor(img, dtype=torch.float32).unsqueeze(0)  # [1, H, W, D]
        
        # Get label
        if session_id not in self.label_map:
            raise ValueError(f"Session ID '{session_id}' not found in CSV")
        
        label_str = self.label_map[session_id]
        if label_str not in self.label_map_int:
            raise ValueError(f"Label '{label_str}' not in label_map_int")
        
        label_int = self.label_map_int[label_str]
        label_tensor = torch.tensor(label_int, dtype=torch.long)
        
        return img_tensor, label_tensor

# -----------------------
# Example usage
# -----------------------
train_dataset = MRIDataset("datasets/train/images", "datasets/train/images/matched_preprocessed_sessions_corrected.csv")
val_dataset   = MRIDataset("datasets/val/images", "datasets/train/images/matched_preprocessed_sessions_corrected.csv")
test_dataset  = MRIDataset("datasets/test/images", "datasets/train/images/matched_preprocessed_sessions_corrected.csv")

train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=2, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=2, shuffle=False)

# Test first batch
images, labels = next(iter(train_loader))
print("Batch images shape:", images.shape)  # [batch_size, 1, H, W, D]
print("Batch labels:", labels)              # tensor([0, 1, ...])
```

```python
import torch
from torch.utils.data import Dataset
import nibabel as nib
import pandas as pd
import os
import numpy as np

class MRIMaskDataset(Dataset):
    def __init__(self, mri_folder, mask_folder, csv_file, label_map):
        self.mri_folder = mri_folder
        self.mask_folder = mask_folder
        self.df = pd.read_csv(csv_file)
        self.label_map = label_map

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        session_id = row['mri_session_id']
        label_name = row['diagnosis']

        # Load MRI
        mri_path = os.path.join(self.mri_folder, f"{session_id}_normalize_resized.nii")
        mri_img = nib.load(mri_path).get_fdata()
        mri_img = (mri_img - np.min(mri_img)) / (np.max(mri_img) - np.min(mri_img))  # normalize
        mri_img = mri_img[np.newaxis, ...]  # add channel dimension

        # Load predicted mask
        mask_path = os.path.join(self.mask_folder, f"{session_id}_predicted_mask.nii.gz")
        mask_img = nib.load(mask_path).get_fdata()
        mask_img = (mask_img > 0).astype(np.float32)  # binary mask
        mask_img = mask_img[np.newaxis, ...]  # add channel dimension

        # Concatenate MRI + mask → 2 channels
        input_tensor = np.concatenate([mri_img, mask_img], axis=0)
        input_tensor = torch.tensor(input_tensor, dtype=torch.float32)

        # Map label to integer
        label = self.label_map[label_name]
        label_tensor = torch.tensor(label, dtype=torch.long)

        return input_tensor, label_tensor
```

```python
import torch.nn as nn

class Simple3DClassifier(nn.Module):
    def __init__(self, in_channels=2, num_classes=3):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv3d(in_channels, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool3d(2),
            nn.Conv3d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool3d(1)
        )
        self.classifier = nn.Linear(32, num_classes)

    def forward(self, x):
        x = self.encoder(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x

model = Simple3DClassifier()
```

```python
import os
import torch
import nibabel as nib
from torch.utils.data import Dataset
import numpy as np
import pandas as pd

class MRISegClassDataset(Dataset):
    def __init__(self, img_dir, label_dir, csv_path, transform=None):
        self.img_dir = img_dir
        self.label_dir = label_dir
        self.data = pd.read_csv(csv_path)
        self.transform = transform

        # Label mapping for classification
        self.label_map = {"Control": 0, "MCI": 1, "Alzheimer": 2}

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        session_id = row['mri_session_id']

        # File paths
        img_path = os.path.join(self.img_dir, f"{session_id}_normalize_resized.nii")
        label_path = os.path.join(self.label_dir, f"{session_id}_mask_resized.nii")

        if not os.path.exists(img_path):
            raise FileNotFoundError(f"❌ MRI not found: {img_path}")
        if not os.path.exists(label_path):
            raise FileNotFoundError(f"❌ Mask not found: {label_path}")

        # Load image and mask
        img = nib.load(img_path).get_fdata()
        mask = nib.load(label_path).get_fdata()

        # Normalize image
        img = (img - np.min(img)) / (np.max(img) - np.min(img) + 1e-8)

        # Convert to tensors
        img_tensor = torch.tensor(img, dtype=torch.float32).unsqueeze(0)  # [1, D, H, W]
        mask_tensor = torch.tensor(mask, dtype=torch.float32).unsqueeze(0)
        class_label = torch.tensor(self.label_map[row['diagnosis']], dtype=torch.long)

        return img_tensor, mask_tensor, class_label
```

```python
import nibabel as nib
import numpy as np
import os
import cv2
import pandas as pd

# === Paths ===
base_dir = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets"
csv_path = os.path.join(base_dir, "binary_labels.csv")
output_dir = os.path.join(base_dir, "vit_slices")
os.makedirs(output_dir, exist_ok=True)

# Load the CSV
df = pd.read_csv(csv_path)

for i, row in df.iterrows():
    nii_path = os.path.join(base_dir, row["filename"])
    label = row["label"]

    try:
        img = nib.load(nii_path).get_fdata()
        img = np.asarray(img, dtype=np.float32)
        
        # Normalize 0–255
        img = (img - np.min(img)) / (np.max(img) - np.min(img))
        img = (img * 255).astype(np.uint8)
        
        # Take the middle slice along z-axis
        mid_slice = img[:, :, img.shape[2] // 2]
        
        # Resize to 224x224 (ViT standard)
        slice_resized = cv2.resize(mid_slice, (224, 224))
        
        # Save as image
        label_dir = os.path.join(output_dir, str(label))
        os.makedirs(label_dir, exist_ok=True)
        
        out_path = os.path.join(label_dir, f"img_{i}.png")
        cv2.imwrite(out_path, slice_resized)
    except Exception as e:
        print(f"⚠️ Error reading {nii_path}: {e}")

print("✅ All slices saved in:", output_dir)
```

```python
import torch
from torch.utils.data import Dataset
from PIL import Image
from transformers import ViTForImageClassification, ViTFeatureExtractor, Trainer, TrainingArguments
import os

# === Load the feature extractor ===
feature_extractor = ViTFeatureExtractor.from_pretrained("google/vit-base-patch16-224")

# === Custom Dataset ===
class BrainDataset(Dataset):
    def __init__(self, root_dir, feature_extractor):
        self.root_dir = root_dir
        self.feature_extractor = feature_extractor
        self.images = []
        self.labels = []

        # Assume folder structure like:
        # dataset/
        #   ├── Healthy/
        #   ├── Diseased/
        for label, subfolder in enumerate(["Healthy", "Diseased"]):
            folder = os.path.join(root_dir, subfolder)
            for file in os.listdir(folder):
                if file.endswith((".png", ".jpg", ".jpeg")):
                    self.images.append(os.path.join(folder, file))
                    self.labels.append(label)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = Image.open(self.images[idx]).convert("RGB")
        encoding = self.feature_extractor(images=image, return_tensors="pt")
        pixel_values = encoding["pixel_values"].squeeze()
        label = torch.tensor(self.labels[idx])
        return {"pixel_values": pixel_values, "labels": label}


# === Load dataset ===
dataset = BrainDataset(r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\vit_slices", feature_extractor)

train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# === Load pretrained model ===
model = ViTForImageClassification.from_pretrained(
    "google/vit-base-patch16-224",
    num_labels=2,
    id2label={0: "Healthy", 1: "Diseased"},
    label2id={"Healthy": 0, "Diseased": 1},
    ignore_mismatched_sizes=True  # fixes classifier layer mismatch
)

# === Training setup ===
training_args = TrainingArguments(
    output_dir="./vit_output",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    save_strategy="epoch",
    logging_dir="./logs",
    logging_steps=10,
    do_eval=True  # replaces evaluation_strategy for older versions
)

# === Trainer ===
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=feature_extractor
)

# === Train ===
trainer.train()
```

```python
import os
import torch
import nibabel as nib
import numpy as np
import pandas as pd
from transformers import ViTForImageClassification, ViTImageProcessor

# === Load model & processor ===
model_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain\vit_output\checkpoint-6" # change if needed
model = ViTForImageClassification.from_pretrained(model_path)
processor = ViTImageProcessor.from_pretrained("google/vit-base-patch16-224")
model.eval()

# === Path to your test set ===
test_dir =r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets\test\images" # path containing .nii files

results = []

for file in os.listdir(test_dir):
    if file.endswith(".nii") or file.endswith(".nii.gz"):
        path = os.path.join(test_dir, file)
        img = nib.load(path).get_fdata()

        # middle slice
        slice_idx = img.shape[2] // 2
        slice_2d = img[:, :, slice_idx]

        # normalize
        slice_norm = (slice_2d - np.min(slice_2d)) / (np.max(slice_2d) - np.min(slice_2d))
        slice_norm = (slice_norm * 255).astype(np.uint8)

        # convert to RGB
        slice_rgb = np.stack([slice_norm] * 3, axis=-1)

        # predict
        inputs = processor(images=slice_rgb, return_tensors="pt")
        with torch.no_grad():
            outputs = model(**inputs)
            probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
            pred = torch.argmax(probs, dim=-1).item()

        label = model.config.id2label[pred]
        confidence = probs[0][pred].item()

        results.append({
            "File": file,
            "Prediction": label,
            "Confidence": round(confidence, 3)
        })

# save results
df = pd.DataFrame(results)
df.to_csv("vit_test_predictions.csv", index=False)
print(df)
```

```python
import os
import shutil
import pandas as pd

# ==== PATHS (adjust if needed) ====
base_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain\datasets"   # your dataset root
train_src = os.path.join(base_path, "train", "images")
val_src = os.path.join(base_path, "val", "images")

# destination folders (final structure)
train_dst = os.path.join(base_path, "train", "images_final")
val_dst = os.path.join(base_path, "val", "images_final")
os.makedirs(train_dst, exist_ok=True)
os.makedirs(val_dst, exist_ok=True)

def move_and_rename_files(src, dst, prefix):
    files = [f for f in os.listdir(src) if f.endswith(".nii") or f.endswith(".nii.gz")]
    data = []
    for i, file in enumerate(files, start=1):
        # Guess label by filename
        fname_lower = file.lower()
        if "healthy" in fname_lower or "normal" in fname_lower:
            label = "Healthy"
            new_name = f"healthy_{i:03d}.nii"
        elif "mci" in fname_lower:
            label = "MCI"
            new_name = f"mci_{i:03d}.nii"
        elif "alz" in fname_lower:
            label = "Alzheimer"
            new_name = f"alz_{i:03d}.nii"
        else:
            label = "Unknown"
            new_name = f"unknown_{i:03d}.nii"

        old_path = os.path.join(src, file)
        new_path = os.path.join(dst, new_name)
        shutil.copy2(old_path, new_path)  # copy instead of move, safe
        data.append({"filename": new_name, "label": label})
    
    # save labels.csv
    df = pd.DataFrame(data)
    csv_path = os.path.join(os.path.dirname(dst), "labels.csv")
    df.to_csv(csv_path, index=False)
    print(f"✅ {len(df)} files processed → {csv_path}")

# Run for both splits
move_and_rename_files(train_src, train_dst, "train")
move_and_rename_files(val_src, val_dst, "val")
```

```python
import os
import pandas as pd
import shutil

# === Paths ===
base_dir = r"C:\Users\Aadya\OneDrive\Desktop\Brain\vit_slices"
healthy_dir = os.path.join(base_dir, "Healthy")
diseased_dir = os.path.join(base_dir, "Diseased")

# === New folder structure ===
output_dir = os.path.join(base_dir, "organized_dataset")
os.makedirs(output_dir, exist_ok=True)

classes = ["Healthy", "MCI", "Alzheimer"]
for cls in classes:
    os.makedirs(os.path.join(output_dir, cls), exist_ok=True)

# === Prepare CSV data ===
data = []

# Process healthy images
if os.path.exists(healthy_dir):
    for img_name in os.listdir(healthy_dir):
        if img_name.endswith(".png"):
            src = os.path.join(healthy_dir, img_name)
            dst = os.path.join(output_dir, "Healthy", img_name)
            shutil.copy(src, dst)
            data.append({"image_name": img_name, "label": "Healthy"})

# Process diseased images
if os.path.exists(diseased_dir):
    diseased_images = [f for f in os.listdir(diseased_dir) if f.endswith(".png")]
    mid_point = len(diseased_images) // 2

    for i, img_name in enumerate(diseased_images):
        src = os.path.join(diseased_dir, img_name)

        # half labeled Alzheimer, half MCI
        if i < mid_point:
            dst = os.path.join(output_dir, "Alzheimer", img_name)
            label = "Alzheimer"
        else:
            dst = os.path.join(output_dir, "MCI", img_name)
            label = "MCI"

        shutil.copy(src, dst)
        data.append({"image_name": img_name, "label": label})

# === Create CSV ===
df = pd.DataFrame(data)
csv_path = os.path.join(output_dir, "labels.csv")
df.to_csv(csv_path, index=False)

print("✅ Dataset organized successfully!")
print(f"CSV file saved at: {csv_path}")
```

```python
import os
import shutil
import pandas as pd
import random
from pathlib import Path

# === Path to your dataset ===
base_dir = Path(r"C:\Users\Aadya\OneDrive\Desktop\Brain\vit_slices")

# === Fix structure first ===
# Move Diseased images from 'A' to 'Diseased' folder if not already there
a_folder = base_dir / "A" / "Diseased"
diseased_folder = base_dir / "Diseased"
diseased_folder.mkdir(exist_ok=True)

if a_folder.exists():
    for img in a_folder.glob("*.png"):
        shutil.move(str(img), diseased_folder / img.name)
    print("✅ Moved images from A/Diseased → vit_slices/Diseased")
else:
    print("⚠️ No A/Diseased folder found, skipping move step.")

# === Ensure all four folders exist ===
for label in ["Healthy", "Diseased", "AL", "MCI"]:
    (base_dir / label).mkdir(exist_ok=True)

# === Create output directory ===
output_dir = base_dir / "organized_dataset"
for split in ["train", "val"]:
    (output_dir / split / "images").mkdir(parents=True, exist_ok=True)

# === Gather all images and labels ===
categories = ["Healthy", "Diseased", "AL", "MCI"]
data = []
for label in categories:
    folder = base_dir / label
    for img in folder.glob("*.png"):
        data.append({"filename": img.name, "label": label, "path": str(img)})

random.shuffle(data)
split_idx = int(0.8 * len(data))
train_data = data[:split_idx]
val_data = data[split_idx:]

def save_split(data_split, split_name):
    split_img_dir = output_dir / split_name / "images"
    rows = []
    for row in data_split:
        dest_path = split_img_dir / row["filename"]
        shutil.copy(row["path"], dest_path)
        rows.append({"filename": row["filename"], "label": row["label"]})
    pd.DataFrame(rows).to_csv(output_dir / split_name / "labels.csv", index=False)
    print(f"{split_name}: {len(rows)} images processed ✅")

save_split(train_data, "train")
save_split(val_data, "val")

print("\n🎯 Dataset ready in vit_slices/organized_dataset/")
```
