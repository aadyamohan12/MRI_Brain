```python
import nibabel as nib
import numpy as np
from scipy.ndimage import zoom
```

```python
brain_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain1\OAS30002_MR_d2345\OAS30002_MR_d2345_normalize.nii.gz"
label_path = r"C:\Users\Aadya\OneDrive\Desktop\Brain1\OAS30002_MR_d2345\aseg_simplified.nii.gz"

# Load brain and label
brain = nib.load(brain_path)
brain_data = brain.get_fdata()

label = nib.load(label_path)
label_data = label.get_fdata()

# Target shape
target_shape = (128, 128, 128)

# Compute zoom factors
zoom_factors_brain = [t / s for t, s in zip(target_shape, brain_data.shape)]
zoom_factors_label = [t / s for t, s in zip(target_shape, label_data.shape)]

# Resize brain (interpolation ok)
brain_resized = zoom(brain_data, zoom_factors_brain, order=3)  # cubic interpolation

# Resize label (use nearest neighbor to keep integer classes)
label_resized = zoom(label_data, zoom_factors_label, order=0)  # nearest-neighbor

# Save resized files
nib.save(nib.Nifti1Image(brain_resized, brain.affine), 
         r"C:\Users\Aadya\OneDrive\Desktop\Brain1\OAS30002_MR_d2345\OAS30002_MR_d2345_normalize.nii.gz")
nib.save(nib.Nifti1Image(label_resized, label.affine), 
           r"C:\Users\Aadya\OneDrive\Desktop\Brain1\OAS30002_MR_d2345\aseg_simplified.nii.gz")

print("✅ Resizing done!")
```

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
import nibabel as nib
import numpy as np
```

```python
class BrainSegDataset(Dataset):
    def __init__(self, image_dir, label_dir, transform=None):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.image_files = sorted(os.listdir(image_dir))
        self.label_files = sorted(os.listdir(label_dir))
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        # Load brain image
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        img = nib.load(img_path).get_fdata()

        # Load mask
        label_path = os.path.join(self.label_dir, self.label_files[idx])
        label = nib.load(label_path).get_fdata()

        # Convert to float32
        img = img.astype(np.float32)
        label = label.astype(np.int64)  # keep as integers for segmentation

        # Add channel dimension (PyTorch expects [C, D, H, W])
        img = np.expand_dims(img, axis=0)

        # Convert to tensors
        img = torch.tensor(img, dtype=torch.float32)
        label = torch.tensor(label, dtype=torch.long)

        return img, label


# Paths
train_img_dir = r"C:\Users\Aadya\OneDrive\Desktop\Brain1\dataset\train\images"
train_lbl_dir = r"C:\Users\Aadya\OneDrive\Desktop\Brain1\dataset\train\labels"

# Dataset and DataLoader
train_dataset = BrainSegDataset(train_img_dir, train_lbl_dir)
train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)

# Test loading one batch
for imgs, masks in train_loader:
    print("Image shape:", imgs.shape)
    print("Mask shape:", masks.shape)
    break
```

```python
import torch
import torch.nn as nn

# Building block: Convolution + BatchNorm + ReLU
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(DoubleConv, self).__init__()
        self.double_conv = nn.Sequential(
            nn.Conv3d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm3d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv3d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm3d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

# U-Net
class UNet3D(nn.Module):
    def __init__(self, in_channels=1, out_channels=2):  # out_channels=2 for binary segmentation
        super(UNet3D, self).__init__()

        # Encoder (downsampling)
        self.enc1 = DoubleConv(in_channels, 32)
        self.pool1 = nn.MaxPool3d(2)
        self.enc2 = DoubleConv(32, 64)
        self.pool2 = nn.MaxPool3d(2)
        self.enc3 = DoubleConv(64, 128)
        self.pool3 = nn.MaxPool3d(2)

        # Bottleneck
        self.bottleneck = DoubleConv(128, 256)

        # Decoder (upsampling)
        self.up3 = nn.ConvTranspose3d(256, 128, kernel_size=2, stride=2)
        self.dec3 = DoubleConv(256, 128)
        self.up2 = nn.ConvTranspose3d(128, 64, kernel_size=2, stride=2)
        self.dec2 = DoubleConv(128, 64)
        self.up1 = nn.ConvTranspose3d(64, 32, kernel_size=2, stride=2)
        self.dec1 = DoubleConv(64, 32)

        # Final output
        self.final_conv = nn.Conv3d(32, out_channels, kernel_size=1)

    def forward(self, x):
        # Encoder
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool1(e1))
        e3 = self.enc3(self.pool2(e2))

        # Bottleneck
        b = self.bottleneck(self.pool3(e3))

        # Decoder with skip connections
        d3 = self.up3(b)
        d3 = torch.cat((d3, e3), dim=1)
        d3 = self.dec3(d3)

        d2 = self.up2(d3)
        d2 = torch.cat((d2, e2), dim=1)
        d2 = self.dec2(d2)

        d1 = self.up1(d2)
        d1 = torch.cat((d1, e1), dim=1)
        d1 = self.dec1(d1)

        return self.final_conv(d1)


# Test the model
model = UNet3D(in_channels=1, out_channels=2)  # out_channels=2 → brain vs background
x = torch.randn(1, 1, 128, 128, 128)  # dummy input
y = model(x)

print("Input shape:", x.shape)
print("Output shape:", y.shape)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import torch.nn.functional as F

# ---------- Dice Loss for binary segmentation ----------
class DiceLoss(nn.Module):
    def __init__(self, smooth=1e-5):
        super(DiceLoss, self).__init__()
        self.smooth = smooth

    def forward(self, preds, targets):
        # preds = [B, C, D, H, W], targets = [B, D, H, W]
        preds = torch.sigmoid(preds)  # convert logits → probabilities (for binary)
        targets = targets.float()     # binary mask as float

        # Flatten
        preds_flat = preds.view(-1)
        targets_flat = targets.view(-1)

        intersection = (preds_flat * targets_flat).sum()
        dice = (2. * intersection + self.smooth) / (preds_flat.sum() + targets_flat.sum() + self.smooth)
        return 1 - dice  # we want to minimize

# ---------- Dataset Loader ----------
train_dataset = BrainSegDataset(
    image_dir=r"dataset/train/images",
    label_dir=r"dataset/train/labels"
)

train_loader = DataLoader(train_dataset, batch_size=1, shuffle=True)

# ---------- Model ----------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UNet3D(in_channels=1, out_channels=1).to(device)  # out_channels=1 for binary

optimizer = optim.Adam(model.parameters(), lr=1e-4)
num_epochs = 5  # adjust later

# ---------- Training Loop ----------
for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0

    for images, masks in train_loader:
        images, masks = images.to(device), masks.to(device)

        # Convert masks to binary (brain vs background)
        masks = (masks > 0).long()  # everything >0 is brain

        # Forward pass
        outputs = model(images)
        loss = DiceLoss()(outputs, masks)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss/len(train_loader):.4f}")

# ---------- Save Model ----------
torch.save(model.state_dict(), "unet3d_brain_binary.pth")
print("✅ Training complete. Model saved as unet3d_brain_binary.pth")
```
